## Paso 1: Identificar dimensiones clave

```python
import json
import psycopg2
from psycopg2.extras import Json

# Dimensiones para análisis de e-commerce
dimensiones_ecommerce = {
    'dim_customer': [
        'customer_id', 'email', 'registration_date',
        'customer_segment', 'total_orders', 'lifetime_value'
    ],
    'dim_product': [
        'product_id', 'sku', 'name', 'category',
        'brand', 'unit_cost', 'current_price'
    ],
    'dim_time': [
        'date_key', 'full_date', 'year', 'quarter',
        'month', 'day_of_week', 'is_weekend', 'is_holiday'
    ],
    'dim_location': [
        'location_id', 'country', 'region', 'city',
        'postal_code', 'timezone'
    ]
}

print("DIMENSIONES IDENTIFICADAS:")
for dim, atributos in dimensiones_ecommerce.items():
    print(f"• {dim}: {', '.join(atributos[:3])}...")
```

```python
# Exportar dimensiones a postgreSQL, encoding UTF-8
import psycopg2
# Configuración de conexión a la base de datos
conn = psycopg2.connect(
	host="localhost",
	database="ecommerce_db",
	user="postgres",
	password="1234",
	port="5432",
)
cursor = conn.cursor()
# Crear tablas para cada dimensión
cursor.execute("""
CREATE TABLE IF NOT EXISTS dim_customer (
    customer_id SERIAL PRIMARY KEY,
	email VARCHAR(255),
	registration_date DATE,
	customer_segment VARCHAR(50),
	total_orders INT,
	lifetime_value DECIMAL(10, 2)
);
""")
cursor.execute("""
CREATE TABLE IF NOT EXISTS dim_product (
	product_id SERIAL PRIMARY KEY,
	sku VARCHAR(50),
	name VARCHAR(255),
	category VARCHAR(100),
	brand VARCHAR(100),
	unit_cost DECIMAL(10, 2),
	current_price DECIMAL(10, 2)
);
""")
cursor.execute("""
CREATE TABLE IF NOT EXISTS dim_time (
	date_key SERIAL PRIMARY KEY,
	full_date DATE,
	year INT,
	quarter INT,
	month INT,
	day_of_week INT,
	is_weekend BOOLEAN,
	is_holiday BOOLEAN
);
""")
cursor.execute("""
CREATE TABLE IF NOT EXISTS dim_location (
	location_id SERIAL PRIMARY KEY,
	country VARCHAR(100),
	region VARCHAR(100),
	city VARCHAR(100),
	postal_code VARCHAR(20),
	timezone VARCHAR(50)
);
""")
conn.commit()
print("Tablas de dimensiones creadas exitosamente.")
cursor.close()
conn.close()
```
